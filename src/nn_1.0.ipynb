# This is a simple first version of a NN

Importing some stuff

```python
import tensorflow as tf
import tensorflow.keras as ks
import pandas as pd
```

```python
votes = pd.read_csv("../data/formatted/swissvotes_dataset_after_1900_utf8.csv", sep=';')
votes.head()
```

```python
votes.describe()
```

For this first version only the influence of a few parties are checked.

```python
list(votes.columns)
parties = ["p_fdp", "p_svp", "p_glp", "p_mitte"]
outcomes = ["volkja_proz"]
```

Unknown party recommendations are dropped. Recommendations other than yes or no are set to neutral. Yes is mapped to 1, no is mapped to -1, and neutral is 0. The data gets read as strings, so it needs to be converted.

```python
votes[parties] = votes[parties].replace(".", pd.NaT)
votes.dropna(subset=parties, inplace=True)
for p in parties:
    votes[p] = votes[p].astype(int)
for o in outcomes:
    votes[o] = votes[o].astype(float)
```

```python
print(votes[parties+outcomes].head())

votes[parties] = votes[parties].replace(2, -1)
votes[parties] = votes[parties].replace([3, 4, 5, 66, 9999], 0)

print(f"before:{len(votes)}")
votes.dropna(subset=parties+outcomes, inplace=True)
print(f"after:{len(votes)}")

# normalize percentage
votes["volkja_proz"] = votes["volkja_proz"].apply(lambda x: x/100)

votes[parties+outcomes].head()
```

Create test and train split

```python
from sklearn.model_selection import train_test_split

in_train, in_test, out_train, out_test = train_test_split(votes[parties], votes[outcomes], test_size=0.8)
```

## NN
The neural net will have 2 hidden layers to start with. 20 and 10 nodes each. The optimizer is gradient descent, and the metrics are mean squared error.

```python
nn = ks.models.Sequential()

nn.add(ks.layers.Dense(20, input_shape=(len(parties), ), activation="sigmoid"))
nn.add(ks.layers.Dense(10, activation="relu"))
nn.add(ks.layers.Dense(len(outcomes), activation="sigmoid"))
nn.compile(optimizer=ks.optimizers.SGD(learning_rate=0.001), loss="mse")
nn.summary()
```

```python
nn.fit(in_train, out_train, batch_size = 32, epochs = 100)
```

```python
out_predict = nn.predict(in_test)
out_comparison = out_test
out_comparison["pred"] = out_predict
print(out_comparison)
```

Using mean squared error on the predicted values

```python
res = 0
for real, pred in zip(out_comparison["volkja_proz"], out_comparison["pred"]):
    res += (real-pred)**2
res /= len(out_test)

print(f"The NN has an error of {res}")
```
