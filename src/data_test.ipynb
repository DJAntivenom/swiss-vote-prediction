```python
import statsmodels.formula.api as smf

sv_df.dropna(subset=["volkja_proz", "p_svp"])
sv_df = sv_df[sv_df["p_svp"] != "."]

sv_df["volkja_proz"] = sv_df["volkja_proz"].astype(float)
sv_df["p_svp"] = sv_df["p_svp"].astype(float)
sv_df = sv_df[sv_df["p_svp"] != 9999]
sv_df = sv_df[sv_df["p_svp"] != 66]
```

```python
result = smf.ols("volkja_proz ~ C(p_svp)", data=sv_df).fit()
result.summary()
```
